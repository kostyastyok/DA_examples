## Анализ данных клиентов сети

**Создаем таблицу с активными клиентами (совершившие хоть одну покупку за последние пол года)**

```python
# Создаем таблицу с активными аккаунтами за последние пол года 

df = result_to_df("""drop table if exists snb_karusel.users_2021""")

df = result_to_df(
"""
create table if not exists snb_karusel.users_2021 ENGINE = Log() as
select *
from DWHD.ACCOUNTS a 
where LOYALTY_CARD_ID in (select distinct zcustid
						  from snb_karusel.checks_headers_karusel chk 
						  where day_partitioned >= today()-180)
""")
```

**Общее кол-во покупателей:**

```python
# Считаем общее кол-во актиынх клиентов за последние пол года

df = result_to_df("""
select count(distinct LOYALTY_CARD_ID) as count_users
from snb_karusel.users_2021
""")

x = df.loc[0, 'count_users']
print(f'Кол-во актиынх пользователей за последние пол года: {x}')
```

**Посмотрим на распределение покупателей по гендерному признаку:**

```python
# Смотрим расределение мужчин и женщин среди активных пользователей 

df = result_to_df("""
select SEX_NM
     , count(distinct LOYALTY_CARD_ID) as user_count
from DWHD.ACCOUNTS a 
where LOYALTY_CARD_ID in (select distinct LOYALTY_CARD_ID
						  from snb_karusel.users_2021)
group by SEX_NM 
order by user_count desc

""")

# Создаем переменные с группами пользователей по гендерному признаку
women_users = df[df['SEX_NM']=='F'].reset_index().loc[0,'user_count'] + df[df['SEX_NM']=='2'].reset_index().loc[0,'user_count']
men_users = df[df['SEX_NM']=='M'].reset_index().loc[0,'user_count'] + df[df['SEX_NM']=='1'].reset_index().loc[0,'user_count']
unknown_users = df[df['SEX_NM']=='None'].reset_index().loc[0,'user_count'] + df[df['SEX_NM']=='U'].reset_index().loc[0,'user_count']
```

```python
print('Кол-во женщин: ' + str(women_users))
print('Кол-во мужчин: ' + str(men_users))
print('Неразмеченные пользователи: ' + str(unknown_users))
```

**Анализ возраста покупателей:**

```python
# Возраст клиента

df = result_to_df(
"""
select LOYALTY_CARD_ID
	 , toDateOrNull(substring(toString(BIRTH_DT),1,10)) as birthday
	 , round((today() - toDateOrNull(substring(toString(BIRTH_DT),1,10))) / 365) as age
from snb_karusel.users_2021
"""
)
```

Отчистим данные от невалидных значений. Удалим строки с пропусками и значения отрицательного возраста.

Приведем столбец с датой к правильному формату.

```python
df['birthday'] = pd.to_datetime(df['birthday'])
df = df.dropna()
df = df[df['age']>0]
df.head()
```

Мы получили значительный выброс по возрасту - 52 года. Вероятно поле даты рождения заполняется по дефолту датой "1970-01-01" (Возможно ошибка в базе). Чтобы получить репрезентативные данные по распределению, уберем строки с этой датой рождения. Данных о более возрастных польлзователях у нас нет.

Так же видим выброс на значении возраста 22, так же отчистим таблицу от даты рождения "2000-01-01". Возможно это стандартная дата при заполнении формы регистрации в приложении, и люди не меняют ее.

```python
df = df[df['birthday'] > '1970-01-01']
df[df['age']==22]['birthday'].value_counts()
df = df[df['birthday']!='2000-01-01']
df.head()
```

**Проанализирум кол-во дней прошедшие с последней покупки:**

```python
# Посмотрим показатели по датам, когда была последняя покупка
df = result_to_df(
"""
select zcustid 
     , min(day_partitioned) as first_day
     , (today() - first_day) as client_days
	 , max(day_partitioned) as last_day
	 , (today() - last_day) as break
from snb_karusel.checks_headers_karusel chk 
where zcustid in (select distinct LOYALTY_CARD_ID
				  from snb_karusel.users_2021)
--and day_partitioned > '2020-01-01'
group by zcustid 
"""
)

df.head()
```

```python
df.hist('break', bins = 20)
```

```python
median_br = df['break'].median()
actual_br = df[df['break']=<60]['break'].median()
```

```python
print(f'Медианное значение кол-ва дней с последней покупки по всем покупателям за пол года: {median_br}')
print(f'Медианное значение кол-ва дней с последней покупки по всем покупателям с учетом оттока: {actual_br}')
```

## Основные показатели в динамике

### По всем пользователям и магазинам: 

```python
# Посмотрим показатели сети по месяцам с 2021-01-01
# Можно отстроить сортировку по активным юзерам/ работающим магазинам на момент 2022-04
df = result_to_df(
"""
select substring(toString(day_partitioned),1,7) as period
	 , count(distinct zcustid) as users_cnt
     , count(zch_uid) as checks
     , checks/users_cnt as frequency
	 , sum(zsale_vat) as rto
	 , sum(zsale_vat - zcst) as fm 
	 , avg(zsale_vat) as mean_cheak
from snb_karusel.checks_headers_karusel chk
where day_partitioned >= '2021-01-01'
and day_partitioned < '2022-05-01'
--and zcustid in (select * from snb_karusel.good_users)
--and plant in (select * from snb_karusel.working_sap_0422)
group by period 
order by period DESC
"""
)

df['period'] = pd.to_datetime(df['period'])
df
```

```python
n = 0
fig, ax = plt.subplots(6,1)
fig.set_size_inches(14, 31)
for x in df.columns[1:]:
    months = mdates.MonthLocator()
    days = mdates.DayLocator()
    timeFmt = mdates.DateFormatter('%Y-%m')

    sns.lineplot(data=df, x="period", y=x,
                     palette = 'bright', ax = ax[n])
    ax[n].set_title(str(x).upper() , fontsize=12)
    ax[n].xaxis.set_major_locator(months)
    ax[n].xaxis.set_major_formatter(timeFmt)
    ax[n].xaxis.set_minor_locator(days)
    
    
    n+=1
plt.show()
```

***По всем данным видим четкий нисходящий тренд по всем основным метрикам.***

## Данные по открытым магазинам

```python
# Посмотрим показатели сети по месяцам с 2021-01-01
# Можно отстроить сортировку по активным юзерам/ работающим магазинам на момент 2022-04
df = result_to_df(
"""
select substring(toString(day_partitioned),1,7) as period
	 , count(distinct zcustid) as users_cnt
     , count(zch_uid) as checks
     , checks/users_cnt as frequency
	 , sum(zsale_vat) as rto
	 , sum(zsale_vat - zcst) as fm 
	 , avg(zsale_vat) as mean_cheak
from snb_karusel.checks_headers_karusel chk
where day_partitioned >= '2021-01-01'
and day_partitioned < '2022-05-01'
--and zcustid in (select * from snb_karusel.good_users)
and plant in (select * from snb_karusel.working_sap_0422)
group by period 
order by period DESC
"""
)

df['period'] = pd.to_datetime(df['period'])
df
```

```python
n = 0
#fig = plt.figure(figsize=(30,60))
fig, ax = plt.subplots(6,1)
fig.set_size_inches(14, 31)
for x in df.columns[1:]:   
    months = mdates.MonthLocator()
    days = mdates.DayLocator()
    timeFmt = mdates.DateFormatter('%Y-%m')

    sns.lineplot(data=df, x="period", y=x,
                     palette = 'bright', ax = ax[n])
    ax[n].set_title(str(x).upper() , fontsize=12)
    ax[n].xaxis.set_major_locator(months)
    ax[n].xaxis.set_major_formatter(timeFmt)
    ax[n].xaxis.set_minor_locator(days)
    ax[n].tick_params(labelsize=8)
 
    n+=1
```

Так же отмечаем нисходящий тренд по числу покупателей и частоте с начала 2021 года, но РТО и FM стабильны. Всплески в декабре 21 и марте 22, обьясняются праздничным ажиотажем)

**Рассмотрим для примера значения РТО более детально:**

```python
# Данные по открытым магазинам

df = result_to_df(
"""
select plant
	 , substring(toString(day_partitioned),1,7) as period
     , count(distinct zcustid) as users
	 , sum(zsale_vat) as rto
	 , sum(zsale_vat - zcst) as fm 
	 , avg(zsale_vat) as mean_cheak
from snb_karusel.checks_headers_karusel chk 
where day_partitioned >= '2021-01-01' 
and day_partitioned <'2022-05-01'
and plant in (select * from snb_karusel.working_sap_0422)
group by plant, 
         period
""")

df['period'] = pd.to_datetime(df['period'])
df
```

```python
fig = plt.figure(figsize=(30,8))
plt.subplot()

months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%Y-%m')

ax = sns.lineplot(data=df, x="period", y="rto", hue = 'plant',
                 palette = 'bright')
plt.title('РТО по магазинам', fontsize=22)
plt.xlabel('Месяц', fontsize=18)
plt.ylabel('РТО', fontsize=18)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
```

Видим, что по действующим магазинам основные показатели находятся на "обычном уровне". Подтверждаем наши гипотезы о относительной стабильности метрик в действующих магазинах, и что низходящий тренд обусловлен закрытием большого кол-ва гипермаркетов.

### Посмотрим на показатели активной группы покупателей 

```python
# Посмотрим показатели сети по месяцам с 2021-01-01
# Можно отстроить сортировку по активным юзерам/ работающим магазинам на момент 2022-04
df = result_to_df(
"""
select substring(toString(day_partitioned),1,7) as period
	 , count(distinct zcustid) as users_cnt
     , count(zch_uid) as checks
     , checks/users_cnt as frequency
	 , sum(zsale_vat) as rto
	 , sum(zsale_vat - zcst) as fm 
	 , avg(zsale_vat) as mean_cheak
from snb_karusel.checks_headers_karusel chk, snb_karusel.zavody_city zc
where day_partitioned >= '2021-01-01'
and day_partitioned < '2022-05-01'
and chk.plant = zc.loc_sap_code
and zcustid in (select * from snb_karusel.good_users)
--and plant in (select * from snb_karusel.working_sap_0422)
group by period
order by period DESC
"""
)

df['period'] = pd.to_datetime(df['period'])
df
```

```python
n = 0
#fig = plt.figure(figsize=(30,60))
fig, ax = plt.subplots(6,1)
fig.set_size_inches(14, 31)
for x in df.columns[1:]:   
    months = mdates.MonthLocator()
    days = mdates.DayLocator()
    timeFmt = mdates.DateFormatter('%Y-%m')

    sns.lineplot(data=df, x="period", y=x,# hue = 'loc_city', 
                     palette = 'bright', ax = ax[n])
    ax[n].set_title(str(x).upper() , fontsize=12)
    ax[n].xaxis.set_major_locator(months)
    ax[n].xaxis.set_major_formatter(timeFmt)
    ax[n].xaxis.set_minor_locator(days)
    ax[n].tick_params(labelsize=8)
 
    n+=1
```

Как видим, из значений ключевых показателей снижается частота(кол-во покупок, совершенные пользователем за месяц), но показатели РТО и FM остаются стабильными, за исключением сезонных всплесков. Падение частоты компенсируется увеличением среднего чека.

Заключаем что на данный момент у гипермаркетов Карусель есть постоянная стабильная аудитория - около 500к уникальных покупателей в месяц.
